# Descrição do Problema

O futebol é um esporte em que o desempenho dos atletas está diretamente ligado à sua condição física. Lesões podem não apenas prejudicar a carreira dos jogadores, mas também impactar o desempenho dos times. Entender o padrão dessas lesões, assim como os efeitos que elas geram no retorno dos jogadores ao campo, é essencial para decisões técnicas, médicas e gerenciais.

O conjunto de dados utilizado neste projeto é o Player Injuries and Team Performance Dataset, que contém informações sobre lesões de jogadores de futebol em oito clubes da Premier League entre 2019 e 2023, bem como o desempenho de seus times. O arquivo está disponível para download e consulta no Kaggle (https://www.kaggle.com/datasets/amritbiswas007/player-injuries-and-team-performance-dataset).

Este projeto tem como objetivo investigar padrões de lesões e seus impactos sobre o desempenho dos jogadores e dos times ao longo das temporadas, considerando diferentes variáveis. Além da análise exploratória, há espaço para desenvolver modelos preditivos que possam antecipar o tempo de afastamento ou prever o desempenho pós-lesão, sendo ferramentas valiosas para clubes e comissões técnicas.

## Hipóteses do Problema

O objetivo é descobrir padrões e correlações que ajudem a responder perguntas como as seguintes hipóteses levantadas com base no dataset:

- Jogadores mais velhos são mais propensos a sofrer lesões e a permanecerem mais tempo afastados dos gramados?

- Certas posições, como zagueiros ou volantes, estão mais propensas a lesões de longa duração?

- Jogadores com posições mais baixas no FIFA tendem a se lesionar com mais frequência?

- Alguns tipos de lesão têm uma duração média maior do que outros?

## Tipo de Problema

O presente trabalho trata de um problema de **regressão supervisionada**, uma vez que a variável-alvo é numérica contínua e o modelo é treinado a partir de dados rotulados. O objetivo é prever a duração da lesão em dias, com base em características dos jogadores lesionados, como idade, posição e tipo de lesão.

A variável-alvo, duração (dias), foi criada na etapa de pré-processamento a partir da diferença entre a data de retorno e a data da lesão. Por se tratar de uma variável contínua, o problema naturalmente se encaixa no contexto de regressão. Essa abordagem pode ser especialmente útil para estimar o tempo de afastamento de jogadores recém-lesionados, contribuindo para o planejamento médico, tático e de escalação das equipes.

## Atributos do Dataset

O dataset Player Injuries and Team Performance contém 656 registros de jogadores lesionados ao longo de diferentes temporadas de oito clubes (Tottenham, Aston Villa, Brighton, Arsenal, Brentford, Everton, Burnley e Manchester United). Cada linha representa uma lesão sofrida por um jogador e suas informações associadas. Como o dataset apresenta 42 atributos, foram reduzidos para 9 para facilitar a análise.

Os principais atributos considerados no projeto são:

* **nome:** nome do jogador lesionado.
* **time:** clube ao qual o jogador pertencia no momento da lesão.
* **posição:** posição em que o jogador atua (ex: goleiro).
* **idade (anos):** idade em anos do jogador no momento da lesão.
* **temporada:** temporada em que a lesão ocorreu (ex: 2019/20).
* **fifa:** pontuação do jogador no jogo Federação Internacional de Futebol Association (FIFA) na temporada correspondente (indicador de performance/talento).
* **lesão:** descrição da lesão sofrida (ex: lesão na panturrilha).
* **dt_lesão:** data em que ocorreu a lesão do jogador.
* **dt_retorno:** data em que o jogador retornou da lesão.

### Média

A média é uma medida de tendência central que representa o valor típico ou o ponto de equilíbrio de um conjunto de dados. É calculada somando-se todos os valores e dividindo-se pelo número total de observações. É sensível a valores extremos (outliers).

```python
#Média dos atributos numéricos do dataset
df.describe().loc['mean']
```

```python
#Cálculo da média apenas das colunas numéricas
medias = df.select_dtypes(include=np.number).mean()

#Criação do gráfico de barras
plt.figure(figsize=(8, 5))
sns.barplot(x=medias.index, y=medias.values)
plt.title('Média dos atributos numéricos')
plt.ylabel('Valor médio')
plt.xticks(rotation=45)
plt.xlabel('Atributos numéricos')
plt.tight_layout()
plt.show()
```

A média está em torno de 27 anos, o que é comum em esportes profissionais. Isso reflete o auge físico e técnico dos jogadores, que geralmente ocorre entre os 25 e 30 anos. A média é próxima de 78 pontos, sugerindo que os jogadores do conjunto de dados são, em geral, de alto nível. Considerando que a escala da FIFA vai de 0 a 100, esse valor indica uma base com jogadores profissionais que atuam em uma liga de elite.

### Desvio Padrão

O desvio padrão é uma medida de dispersão que quantifica a quantidade de variação ou dispersão de um conjunto de valores. Um desvio padrão baixo indica que os pontos de dados tendem a estar próximos da média do conjunto, enquanto um desvio padrão alto indica que os pontos de dados estão espalhados por uma faixa maior de valores. Ele é a raiz quadrada da variância.

```python
#Desvio padrão dos atributos numéricos do dataset
df.describe().loc['std']
```

```python
#Desvios padrão
desvios = df.describe().loc['std']

#Criação do gráfico de barras
plt.figure(figsize=(8, 5))
sns.barplot(x=desvios.index, y=desvios.values)
plt.title('Desvio padrão dos atributos numéricos')
plt.ylabel('Valor do desvio padrão')
plt.xticks(rotation=45)
plt.xlabel('Atributos numéricos')
plt.tight_layout()
plt.show()
```

Ambos os atributos são relativamente concentrados, com desvios padrões baixos em relação às suas escalas (idade vai até 40 aproximadamente e nota do FIFA até 100).

A maior variabilidade observada na nota FIFA indica que a habilidade técnica dos jogadores tende a apresentar mais diferenças entre si do que a idade.

## Histograma

# Pré-Processamento de Dados

Antes de iniciar o pré-processamento, será criado um novo dataset para não sobrescrever os dados originais.

A *feature engineering* é o processo de seleção, transformação e criação de novos atributos a partir de dados brutos para melhorar o desempenho de modelos de aprendizado de máquina.

Em seguida, será criado um novo atributo (variável-alvo), que representará a duração da lesão em dias. A variável duração (dias) é o resultado da subtração entre a data de retorno (Date of return) e a data da lesão (Date of Injury), representando em dias o tempo de afastamento do jogador.

## Tratamento de Valores Nulos

O dataset original não contém valores ausentes. Entretanto, serão feitas mais análises nas colunas dt_lesão e dt_retorno.

```python
#Verificação de missing values do dataset original
df.isnull().sum()
```

```python
#Conversão das colunas de data
df_copia['dt_lesão'] = pd.to_datetime(df_copia['dt_lesão'], errors='coerce')
df_copia['dt_retorno'] = pd.to_datetime(df_copia['dt_retorno'], errors='coerce')

#Verificação de missing values do dataset que será tratado
df_copia.isnull().sum()
```

```python
#Visualização matricial da nulidade do dataset
ms.matrix(df_copia)
```

A análise de valores ausentes revela que as colunas dt_lesão e dt_retorno são as únicas com dados faltantes no conjunto. Ao todo, 9 registros possuem ao menos uma dessas datas ausentes, sendo que 7 deles apresentam ausência simultânea em ambas. As demais colunas estão completas.

Ao converter as datas, há valores ausentes. Isso acontece porque as datas estão despadronizadas, atrapalhando os cálculos de duração (dias), por exemplo:

1.   Meses abreviados misturados com completos: o mês de janeiro aparece como Jan e January, para os meses de junho e julho isso também ocorre.
2.  "Present" ao invés da data.
3.   Datas escritas com vírgula colada no ano (ex: Sep 16,2020 em vez de Sep 16, 2020).
4.   Erro no ano (ex: Mar 7, 0202 em vez de Mar 7, 2020).

Como essa variável é a variável-alvo do problema de regressão, e a ausência de valor inviabiliza seu uso, serão removidos os registros "Present" (lesões ainda em andamento) e padronizadas as outras datas.


```python
#Criando novamente a cópia do dataset original para desfazer a conversão das datas e tratar os problemas
df_tratado = df.copy()

#Remoção das linhas com 'present' na coluna de retorno
df_tratado = df_tratado[~df_tratado['dt_retorno'].astype(str).str.lower().str.contains('present', na=False)]

#Correção das datas com vírgula colada no ano
df_tratado['dt_lesão'] = df_tratado['dt_lesão'].astype(str).str.replace(r',(?=\d)', ', ', regex=True)
df_tratado['dt_retorno'] = df_tratado['dt_retorno'].astype(str).str.replace(r',(?=\d)', ', ', regex=True)

#Correção de erro no ano
df_tratado['dt_lesão'] = df_tratado['dt_lesão'].str.replace('0202', '2020')
df_tratado['dt_retorno'] = df_tratado['dt_retorno'].str.replace('0202', '2020')

#Padronização dos formatos de datas (meses completos e abreviados misturados)
from dateutil import parser

def padronizar_data(data_str):
    try:
        return parser.parse(data_str).strftime('%b %d, %Y')
    except:
        return None  #Retorna None se a conversão falhar

df_tratado['dt_lesão'] = df_tratado['dt_lesão'].apply(padronizar_data)
df_tratado['dt_retorno'] = df_tratado['dt_retorno'].apply(padronizar_data)

#Conversão das colunas de data
df_tratado['dt_lesão'] = pd.to_datetime(df_tratado['dt_lesão'], errors='coerce')
df_tratado['dt_retorno'] = pd.to_datetime(df_tratado['dt_retorno'], errors='coerce')

#Criação da nova coluna 'duração (dias)'
df_tratado['duração (dias)'] = (df_tratado['dt_retorno'] - df_tratado['dt_lesão']).dt.days

#Seleção das 7 primeiras colunas + da nova coluna
df_tratado = df_tratado[['nome', 'time', 'posição', 'idade (anos)', 'temporada', 'fifa', 'lesão', 'duração (dias)']].copy()
df_tratado.head(10)
```

Antes da conversão das colunas de data, foi realizada uma etapa de limpeza e padronização dos valores. Linhas contendo o termo "Present" foram removidas, pois impossibilitam o cálculo da duração da lesão.
Além disso, foram corrigidas datas com vírgulas coladas ao ano, possíveis erros de digitação no ano, e padronizados os nomes dos meses. Essas medidas garantiram que a conversão para o tipo datetime fosse realizada de forma segura e que a variável duração (dias) fosse corretamente calculada.

```python
#Visualização matricial da nulidade do dataset
ms.matrix(df_tratado)
```

Após o tratamento dos dados ausentes, o conjunto final conta com 653 registros completos, sem nenhuma coluna com valores faltantes. Isso garante maior confiabilidade nas análises posteriores.

## Separação em conjuntos de treino e teste

A seguir, será feita a separação das variáveis independentes (x) e da variável-alvo (y), e a divisão dos dados em conjuntos de treino e teste, o que é uma prática comum em problemas supervisionados de regressão.

```python
#Separação dos eixos x e y
x = df_tratado.drop(columns=['duração (dias)'])
y = df_tratado['duração (dias)']
```

Depois disso, é interessante transformar variáveis categóricas. Para aplicar um modelo de regressão, como regressão linear, é necessário transformar as variáveis categóricas (como time, posição, lesão, etc.) usando One-Hot Encoding. Além disso, pode haver a divisão dos dados em treino e teste com o train_test_split do scikit-learn.

```python
#Codificação das variáveis categóricas usando one-hot encoding
x_encoded = pd.get_dummies(x, drop_first=True)

#Divisão dos dados em treino e teste (80% treino, 20% teste)
x_train, x_test, y_train, y_test = train_test_split(
    x_encoded, y, test_size=0.2, random_state=42
)
```

```python
#Verificação das dimensões
print(f"Dimensões de x_train: {x_train.shape}")
print(f"Dimensões de x_test: {x_test.shape}")
print(f"Dimensões de y_train: {y_train.shape}")
print(f"Dimensões de y_test: {y_test.shape}")
```

Após a criação da variável-alvo duração (dias), realizou-se a separação entre as variáveis independentes (atributos explicativos) e a variável dependente. Para que o modelo de regressão pudesse ser aplicado corretamente, as variáveis categóricas foram transformadas por meio de codificação One-Hot Encoding.

Em seguida, os dados foram divididos em conjuntos de treino (80%) e teste (20%). Essa divisão permite avaliar o desempenho do modelo com dados não vistos durante o treinamento.

## Padronização

A padronização é usada para uniformizar dados categóricos ou corrigir problemas de inconsistência textual. Além da padronização das datas, que foi realizada juntamente com a limpeza dos dados, agora será aplicada às colunas posição e lesão.

```python
#Padronização da coluna 'posição' removendo espaços extras nas extremidades e espaços múltiplos internos
df_tratado['posição'] = df_tratado['posição'].astype(str).str.strip()
df_tratado['posição'] = df_tratado['posição'].str.replace(r'\s+', ' ', regex=True)

#Verificação da distribuição relativa após padronização
df_tratado['posição'].value_counts(normalize=True)
```

A padronização das posições é importante para eliminar inconsistências causadas por espaços extras, diferenças de grafia ou formatos variados, garantindo que todas as ocorrências da mesma posição sejam tratadas de forma uniforme. Isso permite análises mais precisas e confiáveis dos dados, como a distribuição proporcional das posições apresentada.

```python
#Padronização os nomes dos tipos de lesão: tudo em minúsculo e sem espaços extras
df_tratado['lesão'] = df_tratado['lesão'].str.strip().str.lower()

#Verificação da distribuição relativa após padronização
df_tratado['lesão'].value_counts(normalize=True)
```
